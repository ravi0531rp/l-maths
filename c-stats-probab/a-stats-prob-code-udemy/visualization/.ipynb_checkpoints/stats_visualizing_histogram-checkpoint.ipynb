### Histograms


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 1000

data = np.exp( np.random.randn(n)/2 )
```

```python
plt.plot(data)
```

```python
plt.plot(data,'s')
```

```python
# number of histogram bins
k = 40

plt.hist(data,bins=k)
plt.show()
```

```python
y,x = np.histogram(data,bins=k)

# bin centers
xx = (x[1:]+x[:-1])/2 # eg, the x contains 

plt.plot(xx,y)
plt.show()
```

```python
np.histogram(data,bins=k)
```

```python
len(np.histogram(data,bins=k))
```
